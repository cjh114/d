```python
D = int(input("물건 구매가를 입력하세요:"))
sale = 0
if D >= 100000:
  sale = 10
else:
  if D >= 50000:
    sale = 7.5
  elif D >= 10000:
    sale = 5
  else:
    sale = 0

print("구매가:{}".format(D))
print("할인율:{}".format(sale))
print("할인금액:{}".format(D*(sale/100)))
print("지불금액:{}".format(D*(100-sale)/100))
```

```python
age = int(input("나이:"))

price = 3000

if age >= 65 or age <= 7:
  price = 0
elif age <= 18:
  price = 1000

print(price)
```

```python
n = int(input("몇개의 숫자를 더할까요? :"))

total = 0

for i in range(n):
  a = int(input("정수입력: "))
  if a%2 == 0:
    total += a

print(total)
```

```python
ft = int(input())

result = 1

for i in range(1, ft+1):
  result *= i

print(result)
```

```python
import random

answer = random.randint(1, 100)

while True:
  D = int(input("예상숫자 :"))
  if D == answer:
    print("정답!")
    break
  else:
    if answer < D:
      print("UP")
    else:
      print("DOWN")
```

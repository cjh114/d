```python
plt.rc('font', family="NanumBarunGothic")
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_temp.index, df_temp['평균기온'], color='red', marker='o')
plt.plot(df_temp.index, df_temp['최저기온'], color='blue', marker='o')
plt.plot(df_temp.index, df_temp['최고기온'], color='green', marker='o')
plt.title('서울시연간기온')
plt.xlabel('연도')
plt.ylabel('기온')
plt.grid()
plt.show()
```
